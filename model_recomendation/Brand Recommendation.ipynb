```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
brand = pd.read_csv('data/brand.csv')
brand
```

```python
pivot_brand = pd.read_csv('data/pivot_brand.csv')
pivot_brand.fillna(0,inplace=True)
```

```python
pivot_locations = pivot_brand.T
pivot_locations
```

# Locations similarities

```python
locations_similarities = cosine_similarity(pivot_locations)
locations_similarities_df = pd.DataFrame(locations_similarities, index = pivot_locations.index, columns = pivot_locations.index)
locations_similarities_df
```

```python
country = "Australia"
similar_countries = locations_similarities_df.loc[country].sort_values(ascending=False)[1:]
similar_countries
```

# Products similarities

```python
pivot_products = pivot_brand.copy()
pivot_products
```

```python
products_similarities = cosine_similarity(pivot_products)
products_similarities_df = pd.DataFrame(products_similarities, index = pivot_products.index, columns = pivot_products.index)
products_similarities_df
```

# Predict similarity based on description (text_based)

```python
products = brand.groupby(['StockCode',"Description"])["Quantity"].sum().reset_index()
products.drop('Quantity',axis=1,inplace=True)
```

```python
products
```

```python
products['Description']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Instantiate the vectorizer object to the vectorizer variable
vectorizer = TfidfVectorizer(min_df=2, max_df=0.7)

# Fit and transform the plot column
vectorized_data = vectorizer.fit_transform(products['Description'])

# Look at the features generated
print(vectorizer.get_feature_names_out() )
```

```python
tfidf_df = pd.DataFrame(vectorized_data.toarray(), columns = vectorizer.get_feature_names_out())
tfidf_df.index =  products["StockCode"]
print(tfidf_df.head())
```

```python
similarities = cosine_similarity(tfidf_df)
cosine_similarity_df = pd.DataFrame(similarities, index =tfidf_df.index, columns=tfidf_df.index)
cosine_similarity_df
```

```python
stockcode = '10123C'
order = cosine_similarity_df.loc[stockcode].sort_values(ascending=False)
order
```
